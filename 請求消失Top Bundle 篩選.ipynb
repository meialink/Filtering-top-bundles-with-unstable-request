```python
df = pd.read_csv(path+"SSP Report for %s~%s.csv" %(datelist[1], datelist[0]))
```

```python
df
```

```python
# Split to day
df4 = df[:500]  #四天前
df3 = df[500:1000] #三天前
df2 = df[1000:1500] #兩天前
df1 = df[1500:] #昨天
```

```python
# Daily Cost
Cost_1 = sum(df1.Cost)
Cost_2 = sum(df2.Cost)
Cost_3 = sum(df3.Cost)
Cost_4 = sum(df4.Cost)
```

```python
bundles = set(df1[:30].loc[:,'Ssp Bud']) | set(df2[:30].loc[:,'Ssp Bud']) | set(df3[:30].loc[:,'Ssp Bud']) | set(df4[:30].loc[:,'Ssp Bud'])
dfb = pd.DataFrame(bundles, columns =['Ssp Bud'])

dfb = dfb.merge(df1.iloc[:,[1,2]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Valid Ssp Request':'Request-1'})
dfb = dfb.merge(df2.iloc[:,[1,2]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Valid Ssp Request':'Request-2'})
dfb = dfb.merge(df3.iloc[:,[1,2]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Valid Ssp Request':'Request-3'})
dfb = dfb.merge(df4.iloc[:,[1,2]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Valid Ssp Request':'Request-4'})

dfb = dfb.merge(df1.iloc[:,[1,3]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Cost':'Cost-1'})
dfb = dfb.merge(df2.iloc[:,[1,3]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Cost':'Cost-2'})
dfb = dfb.merge(df3.iloc[:,[1,3]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Cost':'Cost-3'})
dfb = dfb.merge(df4.iloc[:,[1,3]], on = 'Ssp Bud', how = 'left')
dfb = dfb.rename(columns = {'Cost':'Cost-4'})

dfb['Cost-1 %'] = dfb['Cost-1']/Cost_1
dfb['Cost-2 %'] = dfb['Cost-2']/Cost_2
dfb['Cost-3 %'] = dfb['Cost-3']/Cost_3
dfb['Cost-4 %'] = dfb['Cost-4']/Cost_4

dfb = dfb.fillna(0)

dfb['Req_Delta1'] = (dfb['Request-1']-dfb['Request-2'])/dfb['Request-2']
dfb['Req_Delta2'] = (dfb['Request-2']-dfb['Request-3'])/dfb['Request-3']
dfb['Req_Delta3'] = (dfb['Request-3']-dfb['Request-4'])/dfb['Request-4']

dfb['Cost_Delta1'] = (dfb['Cost-1']-dfb['Cost-2'])/dfb['Cost-2']
dfb['Cost_Delta2'] = (dfb['Cost-2']-dfb['Cost-3'])/dfb['Cost-3']
dfb['Cost_Delta3'] = (dfb['Cost-3']-dfb['Cost-4'])/dfb['Cost-4']
dfb
```

```python
dfb = dfb.loc[(abs(dfb['Req_Delta1']) >= 0.5) | (abs(dfb['Req_Delta2']) >= 0.5) | (abs(dfb['Req_Delta3']) >= 0.5),:]
dfb.rename(columns = {'Ssp Bud':'Bundle'}, inplace=True)
dfb.to_excel(path+'請求消失Top Bundle %s.xlsx' %datelist[0], index=False)
```
